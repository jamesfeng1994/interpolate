```python
from interpolate_class import interpolate
import numpy as np
import matplotlib.pyplot as plt
import ctypes
```

```python
x = [1,2,3,4,5]
y = [0,1,5,9,10]
interpolate_types = ["linear", "cubic"]
extrapolate_types = ["flat", "linear"]
```

```python
fig, axs = plt.subplots(2,2)
fig.set_size_inches(10, 7)

k = 0
for interp_t in interpolate_types:
    for extrap_t in extrapolate_types:
        interpolate_fun = interpolate(np.array(x),np.array(y),interp_t,extrap_t)
        x_test = np.linspace(0,6,50)
        y_test = [interpolate_fun.interp(xx) for xx in x_test]
        axs[k // 2, k % 2].plot(x_test,y_test,'bo')
        axs[k // 2, k % 2].set_title(interp_t + " " + extrap_t)
        k += 1
```
